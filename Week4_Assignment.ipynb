```python
import numpy as np
```

```python
# Sigmoid Activation Function
# Formula: sigmoid(x)= 1 / (1 + exp(-x))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([-3, -1, 0, 1, 3])
print(sigmoid(x))
```

```python
# Tanh Activation Function
# Formula: tanh(x) = (exp(x) - exp(-x)) / (exp(x) + exp(-x))

def tanh(x):
    return np.tanh(x)

x = np.array([-3, -1, 0, 1, 3])
print(tanh(x))
```

```python
# ReLU Activation Function
# Formula: relu(x) = max(0,x)

def relu(x):
    return np.maximum(0, x)

x = np.array([-3, -1, 0, 1, 3])
print(relu(x))
```

```python
# LeakyRelu Activation Function
# Formula: leaky_relu(x) = x if x > 0 else alpha*x (alpha=0.01)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

x = np.array([-3, -1, 0, 1, 3])
print(leaky_relu(x))
```

```python
# Softmax Activation Function
# Formula: softmax(x) = exp(x) / sum(exp(x))

import numpy as np

def softmax(x):
    e = np.exp(x - np.max(x))  # numerical stability
    return e / e.sum()

x = np.array([1.0, 2.0, 3.0])
print(softmax(x))
```
